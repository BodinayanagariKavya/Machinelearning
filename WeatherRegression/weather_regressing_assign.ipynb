```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("C:/Users/hp/Desktop/pandas/weatherHistory.csv")[:10000]

features = np.array(dataset["Temperature (C)"]) 
labels = np.array(dataset["Humidity"])
```

```python
weights = np.random.randn()
bias = np.random.randn()

print(weights,bias)
```

```python
preds = weights*features + bias

plt.scatter(features, labels)
plt.plot(features, preds, color="purple")
plt.show()
```

```python
def mse(preds, labels):
    return np.sum((preds - labels)**2)/len(labels) # mean squared error
loss = mse(preds, labels)
print(loss)
```

```python
# gradient descent
# -(y - y')* x <---- w
# -(y - y')    <---- b

weights_grad = np.sum(-1*(labels - preds) * features)/len(labels)
bias_grad = np.sum(-1*(labels - preds))/len(labels)

print(weights_grad, bias_grad)
```

```python
lr = 0.1

if weights_grad < 0:
    weights += lr
else:
    weights -= lr

if bias_grad < 0:
    bias += lr
else:
    bias -= lr
```

```python
print(weights, bias)
```

```python
for i in range(10000):
    weights_grad = np.sum(-1*(labels - preds) * features)/len(labels)
    bias_grad = np.sum(-1*(labels - preds))/len(labels)

    lr = 0.001

    if weights_grad < 0:
        weights += lr
    else:
        weights -= lr

    if bias_grad < 0:
        bias += lr
    else:
        bias -= lr

    preds = weights*features + bias

    loss = mse(preds, labels)
    print(loss)
```
